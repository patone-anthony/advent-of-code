```python
with open('input 12-11.txt') as f:
#     file = f.read().splitlines() 
    file = f.read().split('\n\n')
```

```python
file[0]
```

```python
# store a list of monkeys
monk_lst = []
for monk in file:
#     print(monk)
    _monk = monk.split('\n')
    monk_lst.append(_monk)
    print(f'\n {_monk}')
```

```python
class Monkey:
    def __init__(self, name, worry, items, oper, op_num, divis, if_true, if_false, inspect):
        self.name = name
        self.worry = worry
        self.items = items
        self.oper = oper
        self.op_num = op_num
        self.divis = divis
        self.if_true = if_true
        self.if_false = if_false
        self.inspect = inspect
```

```python
final = []
for index, monk in enumerate(monk_lst):
#     print(monk)
#     print(index)
    for i in range(len(monk)):
#         print(f'\n i={i}')
        
        # index/name
        if i == 0:
            pass
#             _ind = monk[i].split('Monkey ')[1]
#             index = _ind.split(':')[0]
#             print(index)
        
        #items
        elif i == 1:
#             print(monk[i])
            _items = monk[i].split('Starting items: ')[1]
#             print(f'_itmems = {_items}')
            items = [int(x) for x in _items.split(', ')]
#             print(f'final items= {items}')
            
        # oper:
        elif i == 2:
            #oper (either + or *)
            # locate either operator, store char and position
            # split on position to get number following
            for j in range(len(monk[i])):
                if monk[i][j] == '*' or monk[i][j] == '+':
#                     print(f'char found {monk[i][j]} index={j}')
                    oper = monk[i][j]
                    op_num = monk[i][j+2:]
#                     print(f'op_num {op_num}')
                    break
                    
        # divis
        elif i == 3:
            divis = int(monk[i].split('Test: divisible by ')[1])
#             print(f'divis={divis}')
            
         #if true
        elif i == 4:
            if_true = int(monk[i].split('If true: throw to monkey ')[1])
#             print(f'divis={if_true}')
               
         #if false
        elif i == 5:
            if_false = int(monk[i].split('If false: throw to monkey ')[1])
#             print(f'divis={if_false}')  
        
    temp = Monkey(index, 0, items, oper, op_num, divis, if_true, if_false, 0)
    final.append(temp)

for i in final:
    print(vars(i))
```

```python
# loop through 20 times

# 1. if there are items, iterate through items
# 2. multiply worry level by popped item
# 3. get operator and divid divide that number by 

for i in range(100):
    print(f'\n round={i}')
    for monk in final:
#         print('\n',vars(monk))
        while monk.items:
            monk.inspect += 1
#             print(monk.items)
            popped_item = monk.items.pop(0)
            monk.worry += popped_item
#             print(f'popped {popped_item}, worry: {monk.worry}')

            # get operator (can be + or *)
            # check operation can be add or multiply
            if monk.oper == '*':
                if monk.op_num == 'old':
                    monk.worry *= monk.worry 
                elif monk.op_num.isdigit():
                    monk.worry *= int(monk.op_num)
#                     print(f'monk.worry {monk.worry} * op_num {int(monk.op_num)}')
            elif monk.oper == '+':
                if monk.op_num == 'old':
                    monk.worry += monk.worry          
                elif monk.op_num.isdigit():
                    monk.worry += int(monk.op_num)
#                     print(f'monk.op_num is digit {monk.op_num.isdigit()}')

#             print(f'monk.worry={monk.worry}')
#             monk.worry = int(monk.worry / 3)
#             print(f'monk.worry={monk.worry}')
            # divisible
            
#             print(f'check if monk.worry={monk.worry} % divis={monk.divis}  equals={monk.worry%divis}')
            if monk.worry % monk.divis == 0:
#                 print('if true')
                final[monk.if_true].items.append(monk.worry) 
#                 print(f'{monk.worry} thrown to monkey_{monk.if_true}')
            else:
#                 print('if false')
#                 print(vars(final[if_false]))
                final[monk.if_false].items.append(int(monk.worry))  
#                 print(f'{int(monk.worry)} thrown to monkey_{monk.if_false}')
            print(monk.worry)
            monk.worry = 0
            
for i in final:
    print(vars(i))
```
