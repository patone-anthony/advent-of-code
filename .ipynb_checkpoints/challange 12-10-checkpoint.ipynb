```python
with open('input 12-10.txt') as f:
    file = f.read().splitlines()        
```

```python
lst = []
counter = 0
cycle = 1
x = 1

timer_1 = [0, 0]
timer_2 = [0, 0]

var1 = None
var2 = None


for line in file:
    cmd = line.split()
    print(f'\n{cmd}')
    print(f'cycle: {cycle}')  

    if cmd[0] == 'addx':
        cycle += 1        

    
    # add x needs to happen after execution (at end of cycle)
    if timer_1[0] > 1:
        timer_1[0] -= 1
    elif timer_1[0] == 1:
        print(f'timer_1 finished, incrementing x by {timer_1[1]}')
        x += timer_1[1]
        timer_1[0] = 0
        timer_1[1] = 0

    if timer_2[0] > 1:
        timer_2[0] -= 1
    elif timer_2[0] == 1:
        print(f'timer_2 finished, incrementing x by {timer_2[1]}')
        x += timer_2[1]
        timer_2[0] = 0
        timer_2[1] = 0

    if cmd[0] == 'noop':
        pass
    if cmd[0] == 'addx':
        print('digit found')
        print(f'timer_1={timer_1}, timer_2={timer_2}')
        if not timer_1[0]:
            timer_1[0] = 2
            timer_1[1] = int(cmd[1])
        elif not timer_2[0]:
            print('setting timer 2')
            timer_2[0] = 2
            timer_2[1] = int(cmd[1])
    
    print(f'x={x}')    


    
    print(f'timer_1={timer_1}, timer_2={timer_2}')
 

    if cmd[0] == 'noop':
        cycle += 1

    if cmd[0] == 'addx':
        cycle += 1

    if cycle % 40 == 20:
        lst.append(cycle*x)
        print(f'######## added to list: {cycle} x {x} = {cycle*x}')
   

    
sum(lst)
```
